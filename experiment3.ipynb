```python
#Create a function plotvector() and visualize the transformations
import numpy as np
import matplotlib.pyplot as plt

def plotvector(vec ,trans_matrix):
  #Apply transformation
  trans_vec=np.matmul(trans_matrix, vec)

  #plot inital vector 
  plt.quiver(0, 0, vec[0], vec[1], angles='xy', scale_units='xy' , scale=1, color='green')
  plt.xlim(-10,10)
  plt.ylim(-10,10)
  plt.title("Intial vector")
  plt.show()

  #Plot transformed vector 
  plt.quiver(0 ,0 , trans_vec[0], trans_vec[1], angles='xy', scale_units='xy', scale=1, color='green')
  plt.xlim(-10, 10)
  plt.ylim(-10, 10)
  plt.title("Transformed Vector")
  plt.show()
```

```python
initial_vec =np.array([3, 2])
trans_matrix = np.array([[2,0], [0,2]])
plotvector(initial_vec, trans_matrix)
```

```python
#Create int sample set of size 20 ,Read matrix A and B of size 5X4
#Addition
#subtraction
import numpy as np
sample_set1= np.random.randint(1,101,20)
sample_set2= np.random.randint(1,101,20)
A= sample_set1.reshape((5,4))
B=sample_set2.reshape((5,4))
print('A matrix')
print(A)
print('B matrix')
print(B)
C=np.add(A,B)
print('Addition')
print(C)
D=np.subtract(A,B)
print('Subtraction')
print(D)

```

```python
#Multiplication
import numpy as np
sample_set1= np.random.randint(1,101,20)
sample_set2= np.random.randint(1,101,20)
A= sample_set1.reshape((5,4))
B=sample_set2.reshape((4,5))
print('A matrix')
print(A)
print('B matrix')
print(B)
C=np.matmul(A,B)
print('Multiplication')
print(C)
```

```python
#Divison
import numpy as np
sample_set1= np.random.randint(1,101,20)
sample_set2= np.random.randint(1,101,20)
A= sample_set1.reshape((5,4))
B=sample_set2.reshape((5,4))
print('A matrix')
print(A)
print('B matrix')
print(B)
C=np.divide(A,B)
print('Divison')
print(C)
```
