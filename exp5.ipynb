SYSTEMATIC SAMPLING

```python
def systematic_sampling(df, step):
 
    indexes = np.arange(0, len(df), step=step)
    systematic_sample = df.iloc[indexes]
    return systematic_sample

systematic_sample = systematic_sampling(df, 5)
# print(systematic_sample)
print(systematic_sample.head(50))
```

Cluster sampling

```python
clusters = data['PAYMENTS'].unique()

sample_clusters = random.sample(list(clusters), 2)

# Select all rows that belong to the sample clusters
sample = data.loc[data['PAYMENTS'].isin(sample_clusters)]
sample
```

```python
import scipy.stats as stats

sample_mean = sample['BALANCE'].mean()
pop_mean = df['BALANCE'].mean() # population mean
t_statistic, p_value = stats.ttest_1samp(df['BALANCE'], pop_mean)
print(p_value)
if p_value < 0.05:
    print("The sample mean is significantly different from the population mean.")
else:
    print("The sample mean is not significantly different from the population mean.")
```

```python
# Get the sample mean and standard deviation
sample_mean = sample['BALANCE'].mean()
sample_std = sample['BALANCE'].std()

# Calculate the standard error and margin of error
std_error = sample_std / np.sqrt(100)
margin_error = stats.t.ppf(0.975, df=99) * std_error

# Calculate the confidence interval
ci = (sample_mean - margin_error, sample_mean + margin_error)
print("The 95% confidence interval is:", ci)

print("Population mean: ",df['BALANCE'].mean())
```
